```python
def get_sentence(sent_id, lines):
    sent_lines = [l for l in lines if l.startswith(str(sent_id)+'-')]
    sent_df = []
    for line in sent_lines:
        x = line.split('\t')
        y = {'id': x[0], 'word': x[2], 'constitutive_component': x[8], 'constitutive_policydemic':x[15], 'regulative_component':x[19], 'regulative_policydemic':x[24]}
        senf_df = sent_df.append(y)
    sent = ' '.join([y['word'] for y in sent_df])
    return sent, pd.DataFrame(sent_df)
```

```python
def get_result_sentence(sent_id, lines):
    sent_lines = [l for l in lines if l.startswith(str(sent_id)+'-')]
    sent_df = []
    for line in sent_lines:
        x = line.split('\t')
        y = {'id': x[0], 'word': x[2],'prediction':x[6]}
        senf_df = sent_df.append(y)
    sent = ' '.join([y['word'] for y in sent_df])
    return sent, pd.DataFrame(sent_df)
```

```python
def main_sentence(sent_df):
    reg_comp = sent_df.regulative_component
    reg_comp = [x.split('|') for x in reg_comp]
    while True:
        values = []
        for x in reg_comp:
            values.append(x[0])
        if len(np.unique(values)) == 1 or (len(np.unique(values)) == 2 and '_' in np.unique(values)):
            reg_comp = [x[1:] for x in reg_comp]
            reg_comp =[['_'] if x==[] else x for x in reg_comp]
        else:
            return [r[0].split(')')[0][1:] for r in reg_comp]
```

```python
# sent_id - int(regulative_benchmark[i]
def compare_to_GoldStandard(regulative_sent_id, lines, results_lines):
    sent, sent_df = get_sentence(int(regulative_benchmark[regulative_sent_id]), lines)
    sent_words = sent_df.word
    result_sent, result_sent_df = get_result_sentence(regulative_sent_id+1, results_lines)
    result_sent_df['prediction'] = [x.split(')')[0][1:] for x in result_sent_df['prediction']]
    result_sent_df = result_sent_df.iloc[:-1]
    l = max([len(x.split('|')) for x in sent_df['regulative_component']])
    colnames = ['regulative_component_'+str(i+1) for i in range(l)]
    sent_df_split = pd.DataFrame(columns=['word'] + colnames)
    sent_df_split['word'] = sent_df['word']
    sent_df_split[colnames] = sent_df['regulative_component'].str.split('|', expand=True)
    sent_df_split = sent_df_split.iloc[int(np.where(sent_df_split.word == result_sent_df.word[0])[0])
    :].reset_index()
    for col in colnames:
        sent_df_split[col] = [x.split(')')[0][1:] if x!=None else x for x in sent_df_split[col]]
    return pd.concat([sent_df_split, result_sent_df], axis=1)
```
